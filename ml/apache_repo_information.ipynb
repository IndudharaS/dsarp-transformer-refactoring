```python
import os
import subprocess

BASE_DIR = "repositories"

os.makedirs(BASE_DIR, exist_ok=True)

repositories = [
    ("tika", "https://github.com/apache/tika.git"),
    ("maven", "https://github.com/apache/maven.git"),
    ("camel", "https://github.com/apache/camel.git"),
    ("ant", "https://github.com/apache/ant.git"),
    ("lucene", "https://github.com/apache/lucene.git")
]

for name, url in repositories:

    target = os.path.join(BASE_DIR, name)

    if not os.path.exists(target):

        print(f"Cloning {name}...")

        subprocess.run([
            "git",
            "clone",
            url,
            target
        ])

    else:

        print(f"{name} already exists")
```

```python
import pandas as pd
from git import Repo

rows = []

for name, _ in repositories:

    repo_path = f"repositories/{name}"

    repo = Repo(repo_path)

    commit_count = len(list(repo.iter_commits()))

    rows.append({
        "repository": name,
        "commits": commit_count
    })

df = pd.DataFrame(rows)

df
```
